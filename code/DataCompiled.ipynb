```python
import numpy as npimport matplotlib.pyplot as pltfrom matplotlib import mlabimport csv## read data into arrays of floats by yeardef readData(textData):        infile = open(textData,"r") ## read the data file        content = infile.read()        infile.close()        lines = content.split('\n') ## split up lines        data = []        for each in lines:  ## split each line into separate strings                if len(each) < 1:                        continue          ## get rid of the empty list                                           ##at the end                row = each.split()                temp = {'year':float(row[0]), 'value':float(row[1])}                if temp['year'] >= 1900 and temp['year'] <= 2010:                    data.append(temp)  ##for each data category year, value for the period starting in 1900        return data
```

```python
precipData = readData('totalPrecipGlobalAvg.csv')tempData = readData('avgTempGlobalAvg.csv')co2Data = readData('yrTotCO2.csv')iceData = readData('iceExtSummerAvg.csv')precipRate = readData('rateOfChangeTotalPrecipGlobalAvg.csv')tempRate = readData('rateOfChangeAvgTempGlobalAvg.csv')co2Rate = readData('rateOfChangeTotalCO2emissions.csv')iceRate = readData('rateOfChangeIceExtSummerAvg.csv')
```

```python
data = []rates = []yearVals = []for each in precipData:    yearVals.append(each['year'])    midInt = []for each in precipRate:    midInt.append(each['year'])    ## make an array of years, precipitation, temperature, CO2, ice extent values    for i in range(len(yearVals)):    temp = [yearVals[i], precipData[i]['value'], tempData[i]['value'], co2Data[i]['value'], iceData[i]['value']]    data.append(temp)    ## write all data values into a single csv filewith open('yr_Precip_Temp_CO2_Ice.csv', 'wb') as f:    wtr = csv.writer(f, delimiter= ' ')    wtr.writerows(data)## the file will contain years 1900 - 2010, precipitation in millimiters## temperature in degrees celcius, CO2 emission in millions of metric tons,## and Arctic sea ice extent in square kilometers
```
