```python
# preamble
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline

from pandas import *
```

```python
# define file path and what we want our headers to be, add an annual average column
def pandaRead(variable,year):
    '''Takes data string variables air_temp or precip and years 1901 to 2009'''
    file_path = '../data/'+variable+'Data/'+variable+'.'+str(year)
    # Latitude, longitude in degrees. 
    # Total monthly precip in mm or monthly air temp in Celsius.
    header = ['Longitude', 'Latitude', 'January', 'February', 'March', \
    'April', 'May', 'June', 'July','August', 'September', 'October', \
    'November', 'December']
    df = pd.read_table(file_path,sep='\s*', names = header)
    df['Annual Average'] = df[['January','February', 'March', \
    'April', 'May', 'June', 'July','August', 'September', 'October', \
    'November', 'December']].mean(1)
    return df
```

```python
# Given a range of years, generates filenames for output csv files in "names" 
# and dataframe objects in "frames"

names = ['../data/precipAvg/precip'+str(year)+'.csv' for year in range(2000,2006)]
frames = [pandaRead('precip',year) for year in range(2000,2006)]
```

```python
# Outputs individual csv files of only Longitude, Latitude, and Annual Average of precipitation

for i,frame in enumerate(frames):
    frame.to_csv(names[i],cols=['Longitude','Latitude','Annual Average'])
```

```python
precipAvg = DataFrame(frames[0]['Longitude'])
precipAvg['Latitude'] = frames[0]['Latitude']

for i,year in enumerate(range(2000,2006)):
    precipAvg[str(year)+' Annual Average'] = frames[i]['Annual Average']

precipAvg.to_csv('../data/precipAvg/precip_2000_to_2005.csv')
```

```python
# Note: logColorPlot and linColorPlot functions are deprecated. Use makePlot

# Want one function to: read data for 'precip' or 'air_temp', for a year, choose column to plot, choose log or lin scale.    
def makePlot(variable, column, year, colormap):
    '''Choose 'air_temp' or 'precip' for variable, 'Annual Average' or string of month (capitalize) for column.
    colormap may be log for logarithmic color bar or lin for linear color bar''' 

    data = pandaRead(variable, year)
    
    if variable == 'precip':
        title = column+' precipitation in '+str(year)
    elif variable == 'air_temp':
        title = column+' air temperature in '+str(year)
        
    lat_vals = data['Latitude']
    lon_vals = data['Longitude']
    data_vals = data[column]
    fig = plt.figure(figsize=(12, 6))
    plt.xlabel('Latitude')
    plt.ylabel('Longitude')
    plt.title(title)
    
    if colormap == 'lin':
        plt.scatter(lon_vals, lat_vals, c=data_vals, edgecolors='none')
    elif colormap == 'log':
        plt.scatter(lon_vals, lat_vals, c=data_vals, edgecolors='none',norm=matplotlib.colors.LogNorm())
    
    plt.colorbar()
    plt.show()
    
# colormap latitude/longitude plots - one with a linear color map and one with a logscale color map (to amplify differences for precip)
def logColorPlot(data, column, title):
    '''Takes 'data' as string. Options: precip or air_temp'''
    lat_vals = data['Latitude']
    lon_vals = data['Longitude']
    data_vals = data[column]
    fig = plt.figure(figsize=(12, 6))
    plt.xlabel('Latitude')
    plt.ylabel('Longitude')
    plt.title(title)
    plt.scatter(lon_vals, lat_vals, c=data_vals, edgecolors='none',norm=matplotlib.colors.LogNorm())
    plt.colorbar()
    plt.show()
    
def linColorPlot(data, column, title):
    """Takes 'data' as string. Options: precip or air_temp"""
    lat_vals = data['Latitude']
    lon_vals = data['Longitude']
    data_vals = data[column]
    fig = plt.figure(figsize=(12, 6))
    plt.xlabel('Latitude')
    plt.ylabel('Longitude')
    plt.title(title)
    plt.scatter(lon_vals, lat_vals, c=data_vals, edgecolors='none')
    plt.colorbar()
    plt.show()
```

```python
makePlot('precip','Annual Average',2009,'log')
makePlot('air_temp','Annual Average',2009,'lin')
```

```python
# Check out the difference between the log and linear color maps.
# The linear map washes out differences in low precipitation.
# The log map has a weird lower limit saturation effect, but otherwise is a much richer visualization.
makePlot('precip','Annual Average',1954,'log')
makePlot('precip','Annual Average',1954,'lin')
```
