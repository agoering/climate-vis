```python
import numpy as npimport matplotlib.pyplot as plt## read data into arrays of floats by yeardef readData(textData):        infile = open(textData,"r") ## read the data file        content = infile.read()        infile.close()        lines = content.split('\n') ## split up lines        data = []        for each in lines:  ## split each line into separate strings                if len(each) < 1:                        continue          ## get rid of the empty list                                           ##at the end                temp = each.split()                data.append(temp)    ## data contains lists of data                                          ##for each year         floatData = []        for each in data:                tempLine = []                for element in each:                        temp = float(element)                        tempLine.append(temp)                floatData.append(tempLine)        return floatData
```

```python
CarbonData = readData('CO2ReadData') ## raw data is for carbon emissionsprint CarbonData[0]
```

```python
'''For each year we have the following data:                               Cement       Gas        PerTotal  Gas  Liquids  Solids  Production   Flaring     Capita'''co2data = {}for each in CarbonData:    #co2data[int(each[0])] = each[1:len(each)] ## just to record values in a list    year = int(each[0])    if year < 1950:        co2data[year] = {'total':3.667*each[1], 'gas':3.667*each[2], 'liquids':3.667*each[3], 'solids':3.667*each[4], 'cement':3.667*each[5], 'gas flaring':3.667*each[6]}    else:            co2data[year] = {'total':3.667*each[1], 'gas':3.667*each[2], 'liquids':3.667*each[3], 'solids':3.667*each[4], 'cement':3.667*each[5], 'gas flaring':3.667*each[6], 'per capita':3.667*each[7]}## above the values are in dictionary style with respective categories      print co2data[1954] print co2data[1954]['gas']print co2data[1954]['per capita']
```

```python
total = []gas = []liquids = []solids = []cement = []gasFlaring = []perCapita = []yearVals = []yearValsPerCapita = []years = sorted(co2data.keys())  ## sorting keys in ascending order of yearsfor year in years:    if year < 1950:        total.append(co2data[year]['total'])        gas.append(co2data[year]['gas'])        liquids.append(co2data[year]['liquids'])        solids.append(co2data[year]['solids'])        cement.append(co2data[year]['cement'])        gasFlaring.append(co2data[year]['gas flaring'])        yearVals.append(year)    else:        total.append(co2data[year]['total'])        gas.append(co2data[year]['gas'])        liquids.append(co2data[year]['liquids'])        solids.append(co2data[year]['solids'])        cement.append(co2data[year]['cement'])        gasFlaring.append(co2data[year]['gas flaring'])        perCapita.append(co2data[year]['per capita'])        yearVals.append(year)        yearValsPerCapita.append(year)
```

```python
fig = plt.figure(figsize=(10, 10))plt.xlabel('year')plt.ylabel('CO2 emissions in million of metric tons')plt.title('CO2 emissions from different sources')plt.plot(yearVals, total, label='Total', color='cyan')plt.plot(yearVals, gas, label='Gas', color='blue')plt.plot(yearVals, liquids, label='Liquids', color='green')plt.plot(yearVals, solids, label='Solids', color='red')plt.plot(yearVals, cement, label='Cement Production', color='magenta')plt.plot(yearVals, gasFlaring, label='Gas Flaring', color='yellow')plt.legend(loc='upper left')#plt.show()plt.savefig("CO2-emissions.png")
```
